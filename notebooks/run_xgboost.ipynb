```python
import os
import numpy as np

# ── PATH ─────────────────────────────────────────────
DATA_DIR = "../inara_data/processed"

spectra_path = os.path.join(DATA_DIR, "spectra.npy")
mols_path    = os.path.join(DATA_DIR, "molecules.npy")
ids_path     = os.path.join(DATA_DIR, "system_ids.npy")

# ── LOAD ────────────────────────────────────────────
spectra = np.load(spectra_path)
mols    = np.load(mols_path)
ids     = np.load(ids_path)

MOLECULES = [
    'H2O','CO2','O2','O3','CH4','N2',
    'N2O','CO','SO2','NH3','C2H6','NO2'
]

print("\n=== DATA SHAPES ===")
print("Spectra :", spectra.shape)
print("Molecules:", mols.shape)
print("IDs     :", ids.shape)

# ── GLOBAL CHECK ─────────────────────────────────────
print("\n=== GLOBAL STATS ===")
print("Spectra min/max:", np.min(spectra), np.max(spectra))
print("Molecules min/max:", np.min(mols), np.max(mols))
print("Molecules std (global):", np.std(mols))

# ── CHECK IF ALL ZERO ────────────────────────────────
print("\n=== ZERO CHECK ===")
print("All molecules zero?:", np.all(mols == 0))

# ── PER MOLECULE ANALYSIS ────────────────────────────
print("\n=== PER-MOLECULE STATS ===")

for i, mol in enumerate(MOLECULES):
    y = mols[:, i]

    print(f"\n{mol}:")
    print(f"  unique values (first 10): {np.unique(y)[:10]}")
    print(f"  min={np.min(y):.6f}")
    print(f"  max={np.max(y):.6f}")
    print(f"  mean={np.mean(y):.6f}")
    print(f"  std={np.std(y):.6f}")

# ── CHECK NaNs ───────────────────────────────────────
print("\n=== NaN CHECK ===")
print("NaNs in spectra:", np.isnan(spectra).sum())
print("NaNs in molecules:", np.isnan(mols).sum())

# ── SAMPLE CHECK ─────────────────────────────────────
print("\n=== SAMPLE ROW (first) ===")
print("Molecules[0]:", mols[0])

# ── DISTRIBUTION SUMMARY ─────────────────────────────
print("\n=== VARIANCE SUMMARY ===")
low_var_count = 0

for i, mol in enumerate(MOLECULES):
    std = np.std(mols[:, i])
    if std < 1e-6:
        low_var_count += 1
        print(f"{mol} → LOW VARIANCE ⚠️ (std={std:.8f})")

print(f"\nLow variance molecules: {low_var_count}/{len(MOLECULES)}")

print("\n=== DONE ===")
```
